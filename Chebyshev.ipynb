Chebyshev Polynomials
=====================

```python
import numpy
import matplotlib.pyplot as plt
circle_color = 'blue'
dot_color = 'orange'
nx = 16
ntheta = nx
theta = numpy.linspace(0,numpy.pi,ntheta)
x = numpy.cos(theta)
y = numpy.sqrt(1-x*x)
sizetuple=(16,4)
fig, ax = plt.subplots(ncols=2, figsize = sizetuple)
ax[0].plot(x,y,color=circle_color)
ax[0].plot(x,y,'o',color=dot_color)
ax[0].plot(x,y*0,color=circle_color)

ax[1].plot(x,y,color=circle_color)
ax[1].plot(x,y*0,color=circle_color)

ax[1].plot(x,y,'o', color = dot_color)
ax[1].plot(x,y*0,'o', color=dot_color)

for i in range(nx):
    xi = x[i]
    yi = y[i]
    ax[1].plot([xi,xi],[0,yi],color=dot_color)    
plt.show()
```

```python
n = 32
amp = 0.05

thetah = numpy.linspace(0,numpy.pi,512)
f = amp*numpy.cos(n*thetah)
rpert = 1+f
ypert = rpert*numpy.sin(thetah)
xpert = rpert*numpy.cos(thetah)
xpp = numpy.cos(thetah)


sizetuple=(16,4)
fig, ax = plt.subplots(ncols=2, figsize = sizetuple)
ax[0].plot(x,y,color=circle_color)
ax[0].plot(x,y,'o',color=dot_color)
ax[0].plot(x,y*0,color=circle_color)
ax[0].plot(xpert,ypert, 'r')
#ax[0].plot(xpp,ypert, 'r')

ax[1].plot(xpp,f/amp)
plt.show()
```
